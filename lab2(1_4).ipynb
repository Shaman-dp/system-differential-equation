**Задание 2. Построение фазовых портретов для нелинейных систем**

```
Амплитуда мертвой зоны: 1.4
```

```
#dead zone
def dead_zone_scalar(x, width = 1.4):
    if np.abs(x)<width:
        return 0
    elif x>0:
        return x-width
    else:
        return x+width

dead_zone = np.vectorize(dead_zone_scalar, otypes=[np.float], excluded=['width'])
```

```
#определение функции
def ode(y, t):
    x, y = y
    dydt = [y, np.sign(dead_zone(1 - x)) - y]
    return dydt

#отрисовка фазового портрета системы
drawPhasePortrait(1, 1, -12, 13, -7, 8, ts = 1.5, nt = 301)
```
